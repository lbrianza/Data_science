### Now we define a function that takes a tweet and performs the following actions:
* Remove user handles, hashtags and links
* Remove spaces/punctuation 
* Convert emojs into a word expressing the feeling (e.g 'happy'/'sad')
* Remove stopwords (optional)
* Returns a list containing the cleaned tweet

```python
from nltk.stem.porter import *
stemmer = PorterStemmer()

def text_process(tweet, remove_stopwords = True):

    tweet = tweet.replace('(',' ') #to avoid any unbalanced parenthesis problem
    tweet = tweet.replace(')',' ') #to avoid any unbalanced parenthesis problem
    tweet = tweet.replace('[',' ') #to avoid any unbalanced parenthesis problem
    tweet = tweet.replace(']',' ') #to avoid any unbalanced parenthesis problem
    re2 = re.findall( r"@[\w]*",tweet) #remove user handles
    re3 = re.findall( r"#[\w]*",tweet) #remove hashtags
    re4 = re.findall( r"\shttps?\S+",tweet) #remove urls   
    r = re2 + re3 + re4
    for w in r:
        tweet = re.sub(w,'',tweet)

    emoj = {":)":"happy",
            ":D":"happy",
            ":‑)":"happy",
            ";D":"happy",
            ":(":"sad",
            ":'('":"sad",
            ":‑(":"sad", 
           }
    words = tweet.split()
    new_tweet = [emoj[word] if word in emoj else word for word in words] #replace emojs with 'happy/sad'
    tweet = " ".join(new_tweet)
    tweet = tweet.translate(str.maketrans('', '', string.punctuation)) #remove punctuation
    tweet = "".join(tweet)

    tweet = [stemmer.stem(i) for i in tweet] # stemming
    tweet = ''.join(tweet)

    if (remove_stopwords): #remove stopwords
        return [word.lower() for word in tweet.split() if (word.lower() not in stopwords.words('english'))]
    return [word.lower() for word in tweet.split()]
```

## 1) Simple analysis using Naive Bayes algorithm

First, we will split our dataset into train+test.
Then, we will use TfidfVectorizer to convert all the tweets into sparse vectors using td-idf scores.
Finally, we will train our Naive Bayes model. We will use a pipeline for these last two steps

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
from sklearn.model_selection import train_test_split

msg_train, msg_test, label_train, label_test=train_test_split(df['text'], df['polarity'], 
                                                              test_size=0.2)
```

```python
print(len(msg_train), len(msg_test), len(msg_train) + len(msg_test))
```

### Now we train the Naive Bayes classifier:

```python
from sklearn.pipeline import Pipeline

pipeline = Pipeline([
    ('tfidf', TfidfVectorizer()),  # vectorization + weighted TF-IDF scores
    ('classifier', MultinomialNB()),  # using Naive Bayes classifier
],verbose=True)
```

```python
pipeline.fit(msg_train,label_train)
```

```python
predictions = pipeline.predict(msg_test)
```

```python
len(predictions)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(predictions,label_test))
```

#### We reached about 77% average accuracy, not bad.

Let's use our trained model to predict the sentiment of a few sentences:

```python
def predict(tweet):
    polarity = {0:"Bad", 1:"Good"}
    return polarity[pipeline.predict([tweet])[0]]
```

## 2) Analysis using Convolutional Neural Network

We will use Keras Tokenizer for word embedding, converting the tweets into vectors of integers

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Embedding,LSTM,Flatten,Conv1D,MaxPooling1D,Dropout,GRU
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences
```

```python
X = df['text'].apply(lambda x: text_process(x,remove_stopwords=False)) 
y = df['polarity']

msg_train, msg_test, label_train, label_test=train_test_split(X, y, test_size=0.2, random_state=101)
```

```python
msg_train[1]
```

#### Now we perform the word embedding:

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(msg_train) #fit on our list of tweets and create the dictionary
```

```python
len(tokenizer.word_index)
```

Our vocabolary contains 390,588 words in total

```python
msg_train_emb = tokenizer.texts_to_sequences(msg_train) #embedding: convert tokens into vectors
msg_test_emb = tokenizer.texts_to_sequences(msg_test)
```

```python
msg_train_emb[1]
```

#### We need one additional step: padding, in order to make all vectors of same length

```python
msg_train_emb = pad_sequences(msg_train_emb) #padding: add zeros to make all vectors of same length
```

```python
msg_train_emb[1]
```

```python
maxlen = len(msg_train_emb[1]) #after padding, all messages have same length
maxlen
```

```python
msg_test_emb = pad_sequences(msg_test_emb,maxlen=maxlen)
```

```python
len(msg_test_emb[0])
```

```python
vocab_size = len(tokenizer.word_index) + 1   #0 is reserved
```

#### Now we create our CNN:

```python
model = Sequential()
model.add(Embedding(input_dim = vocab_size, output_dim=32, input_length=maxlen))
model.add(Conv1D(filters=32, kernel_size=8, activation='relu'))
model.add(MaxPooling1D(pool_size=2))
#model.add(Dropout(0.5)) #given the large number of samples, dropout layers don't give much improvement in performance
model.add(Flatten())
model.add(Dense(10, activation='relu'))
#model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])

print(model.summary())
```

#### We add an early stopping to avoid overfitting

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

```python
model.fit(msg_train_emb, label_train, epochs = 10, batch_size=128, verbose = 1,
                    validation_data=(msg_test_emb,label_test),
                   callbacks=[early_stop])
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses[['loss','val_loss']].plot()
```

```python
losses[['acc','val_acc']].plot()
```

```python
predictions = model.predict_classes(np.reshape(msg_test_emb,[-1,maxlen]))
```

```python
print(classification_report(predictions,label_test))
```

```python
model.save('sentiment_analysis_140_cnn.h5', overwrite=True)
```

### We slightly improved and reached 78-79% accuracy, but not dramatically (especially if we consider the different computational time)

# 4) Analysis using RNN with LSTM + word embedding with Word2Vec

In this last step we will use Word2vec for word embedding. Then we will create a weight matrix of the embedded vectors, and we will use it inside the embedding layer in our RNN.

First, we train the Word2Vec model and create the vocabulary of our dataset:

```python
from gensim.models import Word2Vec

model3 = Word2Vec(msg_train, size=100, min_count=1, window=5) # train word2vec model

words = list(model3.wv.vocab)
print(len(words))
```

How does a word look like in this vector space?

```python
print(model3['love']) #vector representation of the word 'love'
```

```python
filename = 'embedding_word2vec.txt'
model3.wv.save_word2vec_format(filename, binary=False)
```

After saving the vocabulary, we create two functions: one to load the embedding from the vocabulary, one to create a matrix containing the weights of each word:

```python
def load_embedding(filename):
    file = open(filename,'r') 	# load embedding into memory
    lines = file.readlines()[1:]
    file.close()
    # create a map of words to vectors
    embedding = dict()
    for line in lines:
        parts = line.split() # key is string word, value is numpy array for vector
        embedding[parts[0]] = np.asarray(parts[1:], dtype='float32')
    return embedding
```

```python
def get_weight_matrix(embedding, vocab):
    vocab_size = len(vocab) + 1 # total vocabulary size plus 0 for padding
    weight_matrix = np.zeros((vocab_size, 100)) # define weight matrix with all 0
    for word, i in vocab.items(): # step vocab, store vectors using the Tokenizer's integer mapping
        weight_matrix[i] = embedding.get(word)
    return weight_matrix
```

#### Finally we convert all tweets into vectors:

```python
tokenizer = Tokenizer()
tokenizer.fit_on_texts(msg_train)
# load embedding from file
raw_embedding = load_embedding('embedding_word2vec.txt')
# get vectors in the right order
embedding_vectors = get_weight_matrix(raw_embedding, tokenizer.word_index)
vocab_size = len(tokenizer.word_index) + 1
print(vocab_size)
```

```python
embedding_vectors.shape
```

### We are now ready to train our RNN model:

```python
model4 = Sequential()
model4.add(Embedding(input_dim = vocab_size, output_dim=model3.wv.vector_size, 
                     input_length=maxlen, 
                     weights=[embedding_vectors], trainable=False))
model4.add(LSTM(128))
model4.add(Flatten())
model4.add(Dense(10, activation='relu'))
model4.add(Dense(1, activation='sigmoid'))

model4.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])

print(model4.summary())
```

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor='val_loss',patience=2)
```

```python
model4.fit(msg_train_emb, label_train, epochs = 10, batch_size=128, verbose = 1,
                    validation_data=(msg_test_emb,label_test),
                   callbacks=[early_stop])
```

```python
model4.save('sentiment_analysis_140_word2vec_lstm.h5', overwrite=True)
```

```python
predictions = model4.predict_classes(np.reshape(msg_test_emb,[-1,maxlen]))
```

```python
print(classification_report(predictions,label_test))
```

### We reached a 83% accuracy: that's the best result, and thanks to Word2Vec we saved a lot of computatonal time.

#### Now we define a function that given a tweet in input, it predicts the probability (>0.5 = positive tweet) and the sentiment of the tweet using our trained model:

```python
def predict(tweet):
    polarity = {0:"Bad", 1:"Good"}
    tweet = text_process(tweet,remove_stopwords=False)
    tweet_emb = tokenizer.texts_to_sequences([tweet])
    tweet_emb = pad_sequences(tweet_emb,maxlen=maxlen)
    print(f"Probability: {model4.predict_proba(np.reshape(tweet_emb,[-1,maxlen]))[0][0]}")
    return polarity[model4.predict_classes(np.reshape(tweet_emb,[-1,maxlen]))[0][0]]
```
